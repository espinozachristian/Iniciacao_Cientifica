```python
'''
    Extrai os municipios da tabela do wikipedia.
    Será utilizado posteriormente para verificação junto aos sites extraidos e checagem.
'''
source_code = requests.get(str('https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_do_Brasil_por_popula%C3%A7%C3%A3o'))
http_return = source_code.status_code
plain_text = source_code.text
soup = BeautifulSoup(plain_text,"html.parser")
table = soup.find("table",class_="wikitable")
lista = []
for items in table.find_all("tr")[1:]:
    data = [' '.join(item.text.split()) for item in items.find_all(['th','td'])]
    municipios = data[2]
    estado = data[3]
    populacao = data[4]
    lista.append([municipios,estado,populacao])
data = pd.DataFrame(lista)
data.rename({0:'Municipio',1:'Estado',2:'Populacao'},inplace=True,axis=1)
writer = pd.ExcelWriter('Wikipedia_estados.xlsx', engine='xlsxwriter')
data.to_excel(writer, sheet_name='Pagina 1')
writer.save()
```

```python
'''
    Devido ao google ter uma ferramenta de bloquear numero grande de Requests,Ocorreu a tentativa abaixo de colocar um delay entre as pesquisas para simular uma pesquisa mais
    humanizada.
    variavel delay_busca que está entre 5 a 11s para cada site buscado.
    variavel delay_pesquisa de 3min a 3min 40s de delay entre cada nova pesquisa de um novo estado.
    Utilizamos a biblioteca from googlesearch import search onde o mesmo busca todos resultados até sua ultima pagina antes de alterar a pesquisa de forma mais automatica.
    dentro do nosso if '.gov' in result: apenas me retorna sites no qual contenha .gov em seu link buscando de primeiro momento esse filtro.
'''
delay_busca = [ 5, 6, 7, 8,9,10]
delay_busca = np.random.choice(delay_busca)
delay_pesquisa = [180,200,220]
delay_pesquisa = np.random.choice(delay_pesquisa)
lista_de_sites = []
for regiao in lista_de_estados:#Entramos na Lista dos estados  [sudeste,sul,nordeste,norte,centro_oeste]
    for estado,abv in regiao.items():#Acessamos dentro de cada lista seu municipio 
        response = search("Prefeituras municipais de" + estado,pause=2)#Iniciamos pela Lib a pesquisa com pause de 2 segundos a cada pesquisa
        for result in response:
            if str(abv) in result:
                try:#Verifico se o site que acabou de ser captado é acessivel
                    url1 = result
                    source_code = requests.get(str(url1))
                    http_return = source_code.status_code
                    plain_text = source_code.text
                    soup = BeautifulSoup(plain_text,"html.parser")
                except:
                    pass
                try:#Pego o Titulo do site que acabou de acessar.
                    title = soup.title.string
                except:
                    title = 'Null'
                time.sleep(delay_busca)#Adicionando o Delay
                print(result ,"\nTitulo Pagina:",title,'\nRetorno:',http_return)
                lista_de_sites.append([result,title,http_return])
        print("##### FIM DE BUSCA...ALTERANDO ESTADO... ####")
        links = pd.DataFrame(lista_de_sites)
        links.rename({0:'Sites',1:'Estado',2:'Resposta Site'},inplace=True,axis=1)
        links.drop_duplicates(subset ="Sites",keep = False, inplace = True)
        links.to_excel('sites_'+estado+'2.xlsx')
        time.sleep(delay_pesquisa)
```
