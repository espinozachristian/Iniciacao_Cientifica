```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
links_with_text = []
def get_link_page(url):
    #print(url)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text,"html.parser")
    for a in soup.find_all('a'):
        #print(a)
        if a.text: 
            link = a.get('href')[7:].split('&')
            if '.gov' in link[0]:
                if 'http' in link[0]:
                        print(link[0])
                        links_with_text.append(link[0])
                        
```

```python
count = 0
while count != 100000:
    url = "http://www.google.com/search?q="+'prefeitura municipal'+'&start='+str(count)
    get_link_page(url)
    count+=10
    print('pagina atual:',count/10)
```

```python
len(links_with_text)
```

```python
links = pd.DataFrame(links_with_text) 
```

```python
links.rename({0:'Sites'},inplace=True,axis=1)
```

```python
links.head(5)
```

```python
links.drop_duplicates(subset ="Sites", 
                     keep = False, inplace = True) 
```

```python
links.to_csv('Sites', sep='\t', encoding='utf-8')
```
