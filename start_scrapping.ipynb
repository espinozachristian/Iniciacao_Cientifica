```python
'''
    Devido ao google ter uma ferramenta de bloquear numero grande de Requests,Ocorreu a tentativa abaixo de colocar um delay entre as pesquisas para simular uma pesquisa mais
    humanizada.
    variavel delay_busca que está entre 5 a 11s para cada site buscado.
    variavel delay_pesquisa de 3min a 3min 40s de delay entre cada nova pesquisa de um novo estado.
    Utilizamos a biblioteca from googlesearch import search onde o mesmo busca todos resultados até sua ultima pagina antes de alterar a pesquisa de forma mais automatica.
    dentro do nosso if '.gov' in result: apenas me retorna sites no qual contenha .gov em seu link buscando de primeiro momento esse filtro.
'''
delay_busca = [ 5, 6, 7, 8,9,10,11]
delay_busca = np.random.choice(delay_busca)
delay_pesquisa = [180,200,220]
delay_pesquisa = np.random.choice(delay_pesquisa)
lista_de_sites = []
for regiao in lista_de_estados:
    for estado in regiao:
        response = search("Prefeituras municipais " + estado,pause=2)
        for result in response:
            if '.gov' in result:
                text = re.split('\\b.\\b',result)[2]
                time.sleep(delay_busca)
                #links_with_text.append([result,text.upper()])
                print(result ," ##Estado: ",text)
                lista_de_sites.append([result,text])
        print("##### FIM DE BUSCA...ALTERANDO ESTADO... ####")
        links = pd.DataFrame(lista_de_sites)
        links.rename({0:'Sites',1:'Estado'},inplace=True,axis=1)
        links.drop_duplicates(subset ="Sites",keep = False, inplace = True)
        links.to_excel('sites_'+estado+'.xlsx')
        time.sleep(delay_pesquisa)
```
